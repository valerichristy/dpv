```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=(iris.feature_names))
iris_df['target'] = pd.Series(iris.target)
iris_df
```

```python
iris_df['target']= iris_df['target'].replace([0,1,2], ['setosa', 'versicolor', 'virginica'])
```

```python
ax = pd.plotting.scatter_matrix(iris_df.iloc[:,0:4], alpha=0.4, figsize=(7,7), diagonal='hist', hist_kwds={'bins':10, 'color':'orange', 'edgecolor':'#FFFFFF'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
sns.pairplot(iris_df, diag_kind="kde", hue='target',markers=["o", "s", "D"])
plt.legend=iris.target_names
```
