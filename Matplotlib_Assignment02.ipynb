```python
import numpy as np
import pandas as pd
from sklearn import datasets

# Loading wine dataset
wine = datasets.load_wine()

# Storing data in DataFrame
df_wine = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df_wine['target'] = wine.target

# Describing statistical data
df_wine.describe()

# Replace target value
df_wine['target']= df_wine['target'].replace([0,1,2], ['class-0', 'class-1', 'class-2'])
```

```python
df_wine
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

# Computing pairwise correlation of columns
corr = df_wine[df_wine.columns].corr()

# Showing the correlation matrix in a heatmap
# Attributes 'flavanoids' and 'od280/od315_of_diluted_wines' are highly correlated with 'target'
rcParams['figure.figsize'] = 12, 9
sns.heatmap(corr, cmap="RdBu", annot = True);
```

```python
plt.figure(figsize=(8,6))
a = sns.scatterplot(df_wine, x='flavanoids', y='od280/od315_of_diluted_wines', style='target', markers=["o", "s", "D"], hue="target", palette="deep", alpha=0.8)
plt.legend(frameon=True, loc='lower right', ncol=3)
plt.annotate('outlier', xy=(5.1, 3.68), xytext=(4.5, 3.2), fontsize=7,
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2", lw=0.3))
```
